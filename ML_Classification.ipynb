```python
import seaborn as sns
data=sns.load_dataset("Iris")
data
```

```python
data['species'].unique()
```

```python
data['species'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
data['species']=a.fit_transform(data['species'])
data
```

```python
data.isnull().sum()
```

```python
x=data.drop(['species'],axis=1)
print(x)
```

```python
y=data['species']
print(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(xtrain,ytrain)
```

```python
accuracy=model1.score(xtest,ytest)*100
print(accuracy)
```

```python
predict=model1.score(xtrain,ytrain)*100
print(predict)
```

```python
v=[[5.1,3.5,1.4,0.2]]
out=model1.predict(v)
print(out)
```

```python
from sklearn.tree import DecisionTreeClassifier
model2=DecisionTreeClassifier()
model2.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
b2=model2.predict(xtest)
a2=accuracy_score(ytest,b2)*100
print(a2)
```

```python
b21=model2.predict(xtrain)
p2=accuracy_score(ytrain,b21)*100
print(p2)
   
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(30,15))
tree.plot_tree(model2,feature_names=x.columns.tolist(),filled=True,max_depth=3)
plt.show
```
